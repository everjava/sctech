```python
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import seaborn as sns  # Biblioteca de visualização que possui datasets de estudo embutidos
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = sns.load_dataset('iris')
```

```python
df.info()
```

```python
X_estudo = df[['sepal_width','sepal_length','petal_width','petal_length']]
y_estudo = df['species']

X_treino, X_teste, y_treino, y_teste = train_test_split(X_estudo, y_estudo, test_size=0.20, random_state=42)
```

```python
# Vamos usar o algoritmo de Árvore de Decisão
modelo_final = DecisionTreeClassifier(max_depth=3)
modelo_final.fit(X_treino, y_treino) # A MÁGICA ACONTECE NO FIT!
print("IA Treinada com sucesso!")
```

```python
modelo_final.fit(X_treino, y_treino)

# 4. Predições
y_pred = modelo_final.predict(X_teste)

from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris.data, iris.target
# 5. Métricas de Avaliação
print("Acurácia:", accuracy_score(y_teste, y_pred))
print("\nMatriz de Confusão:\n", confusion_matrix(y_teste, y_pred))
print("\nRelatório de Classificação:\n", classification_report(y_teste, y_pred, target_names=iris.target_names))
```
